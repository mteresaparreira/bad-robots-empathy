```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
dataset_path = '../../data/allParticipant_data/allParticipants_5fps_downsampled_preprocessed_norm.csv'
superBAD_df = pd.read_csv(dataset_path)
superBAD_df
```

### Train Test split

```python
    # # Extract features and labels

    # # for naive & naive_n datasets
    # features = df.iloc[:, 3:]
    # target_class = df['class'].values

    # # for full & full_n datasets
    features = superBAD_df.iloc[:, 4:]
    target_class = superBAD_df.iloc[:, 2].values
    target_class = target_class.astype('int')
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target_class, test_size = 0.20, random_state = 42)
```

### LogisticRegression

```python
log_classifier = LogisticRegression(max_iter=200, multi_class = 'multinomial', random_state = 42)

log_classifier.fit(X_train, y_train)

y_predict = log_classifier.predict(X_test)
print(classification_report(y_test, y_predict))
```

### GridSearch + LogisticRegression

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200, 300, 400, 500],
    'penalty': ['l1', 'l2'],
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
    'multi_class' : ['ovr', 'multinomial']
}

grid_search = GridSearchCV(LogisticRegression(random_state = 42, multi_class='ovr'), param_grid=param_grid,  refit= True)
grid_search.fit(X_train, y_train)
```

```python
# print best parameter after tuning
print("Best Parameters: ", grid_search.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid_search.best_estimator_)
```

```python
grid_predictions = grid_search.predict(X_test)

# print classification report
print(classification_report(y_test, grid_predictions))
```
