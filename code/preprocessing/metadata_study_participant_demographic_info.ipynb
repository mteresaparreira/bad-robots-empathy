```python
import os
import numpy as np
import pandas as pd
```

Obtain the final_participant IDs who are considered in the study

```python
participant_survey_logs_path = '../../data/survey_log/'

final_participant_list_path = participant_survey_logs_path + 'participant_log.xlsx'
final_participants = []

final_participant_log_df = pd.read_excel(final_participant_list_path)
for indx, row in final_participant_log_df.iterrows():
    if row['Include as Study Data?'] == 'Y':
        final_participants.append(str(int(row['Participant'])))
```

```python
print(f'Number of Participants Considered for the study = {len(final_participants)}')
```

```python
qualtrics_participant_data_path = participant_survey_logs_path + 'Bad Robots Empathy_May 20, 2023_14.26.csv'
qualtrics_participant_survey_metadata_df = pd.read_csv(qualtrics_participant_data_path)

final_participants_survey_df = qualtrics_participant_survey_metadata_df.copy()

for indx, row in final_participants_survey_df[2: ].iterrows():
    if row['randomID'] not in final_participants:
        final_participants_survey_df.drop(indx, inplace=True)
final_participants_survey_df.reset_index(drop=True, inplace=True)
final_participants_survey_df
```

Obtain demographic information of the final participants that are considered in the study, such as  

- Age
- Gender
- Education Level etc

```python
## Calculate mean age and stdev & the participant genders

participant_ages = {
    'Male': [],
    'Female': [],
    'Others': []
}

all_participant_ages = []

for indx, row in final_participants_survey_df[2:].iterrows():
    # Check what gender they identify as and the age correspondingly
    if row['Q11.4'] == 'Male':
        participant_ages['Male'].append(int(row['Q726']))
    elif row['Q11.4'] == 'Female':
        participant_ages['Female'].append(int(row['Q726']))
    else:
        participant_ages['Others'].append(int(row['Q726']))

for key in participant_ages.keys():
    all_participant_ages.append(participant_ages[key])
    print(f'Total Number of {key} Participants = ', len(participant_ages[key]))

print('--------------------')

for key in participant_ages.keys():
    print(f'{key} Participants Maximum Age  =', np.array(participant_ages[key]).max() if participant_ages[key] else None)
    print(f'{key} Participants Minimum Age  =', np.array(participant_ages[key]).min() if participant_ages[key] else None)
    print(f'{key} Participants Age Mean =', np.array(participant_ages[key]).mean() if participant_ages[key] else None)
    print(f'{key} Participants Age Std =', np.array(participant_ages[key]).std() if participant_ages[key] else None)
    print('--------------------')

all_participant_ages = np.concatenate(list(participant_ages.values()))
all_participant_ages_mean = all_participant_ages.mean()
all_participant_ages_std = all_participant_ages.std()

print(f'Overall Participants Ages - Mean = {all_participant_ages_mean}')
print(f'Overall Participants Ages - Std = {all_participant_ages_std}')
```

```python
## Calculate the number of different ethnicities people have been included in the study
participant_ethnicities_df = final_participants_survey_df[final_participants_survey_df['Q11.1'] != 'What is your race/ethnicity? (check all that apply) - Selected Choice']
participant_ethnicities_df = participant_ethnicities_df[participant_ethnicities_df['Q11.1'] != '{"ImportId":"QID53"}']
participant_ethnicities_counts = participant_ethnicities_df['Q11.1'].value_counts().to_dict()

print(f'Total Number of Ethnicities included in the study = {len(participant_ethnicities_counts)}')
print(f'\nEthnicities included in the study :')
participant_ethnicities_counts
```

```python
## Calculate the number of different nationalities people have been included in the study

participant_nationalities = {}

for indx, row in final_participants_survey_df[2:].iterrows():
    key = row['Q727'].strip().lower()
    participant_nationalities[key] = participant_nationalities.get(key, 0) + 1

print(f'Total Number of Nationalities included in the study = {len(participant_nationalities)}')
print(f'\nNationalities included in the study :')
participant_nationalities
```

```python
# Count the occurrences of each education level and store them in a dictionary
# Remove the first row with the question description
education_df = final_participants_survey_df[final_participants_survey_df['Q11.3'] != 'What is your highest level of education?']
education_df = education_df[education_df['Q11.3'] != '{"ImportId":"QID55"}']
education_counts = education_df['Q11.3'].value_counts().to_dict()

# Print the dictionary
education_counts
```

```python
## English proficiency data

english_proficiency_df = final_participants_survey_df[final_participants_survey_df['Q744'] != 'What is your level of English proficiency?']
english_proficiency_df = english_proficiency_df[english_proficiency_df['Q744'] != '{"ImportId":"QID748"}']
english_proficiency_counts = english_proficiency_df['Q744'].value_counts().to_dict()

english_proficiency_counts
```

```python
## Calculate the avg and mean duration taken by the participants to complete the survey

participant_survey_duration = {}

for indx, row in final_participants_survey_df[2: ].iterrows():
#     print(f"Participant ID: {row['randomID']} = {row['Duration (in seconds)']} seconds")
#     print(f"{divmod(int(row['Duration (in seconds)']), 60)}")
    participant_survey_duration[row['randomID']] = int(row['Duration (in seconds)'])

survey_durations = np.array(list(participant_survey_duration.values()))
survey_duration_mean = survey_durations.mean()
survey_duration_mean_mins, survey_duration_mean_secs = divmod(survey_duration_mean, 60)

survey_duration_std = survey_durations.std()
survey_duration_std_mins, survey_duration_std_secs = divmod(survey_duration_std, 60)

survey_duration_max = divmod(survey_durations.max(), 60)
survey_duration_min = divmod(survey_durations.min(), 60)

print(f"Participant Survey Duration Average = {survey_duration_mean_mins}m {survey_duration_mean_secs}s")
print(f"Participant Survey Duration Std = {survey_duration_std_mins}m {survey_duration_std_secs}s")
print(f"Participant Survey Duration Maximum = {survey_duration_max[0]}m {survey_duration_max[1]}s")
print(f"Participant Survey Duration Minimum = {survey_duration_min[0]}m {survey_duration_min[1]}s")
participant_survey_duration
```
